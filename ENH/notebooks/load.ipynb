El hogar, la vivienda y los integrantes del hogar.

```python
import pandas as pd
import missingno as msno
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Original Data

```python
file = '../dataframes/enh.csv'
df = pd.read_csv(file)
df.head()
```

```python
df.replace(to_replace=-1, value=np.NaN, inplace=True)
```

```python
df.shape
```

```python
missingdata = df.columns[df.isnull().any()].tolist()
len(missingdata)
```

```python
msno.matrix(df[missingdata])
plt.savefig('../figures/matrix_original_data.svg')
plt.savefig('../figures/heatmap_original_data.png')
```

```python
msno.bar(df[missingdata], log=False)
plt.savefig('../figures/bar_original_data.svg')
plt.savefig('../figures/heatmap_original_data.png')
```

```python
msno.heatmap(df)
plt.savefig('../figures/heatmap_original_data.svg')
plt.savefig('../figures/heatmap_original_data.png')
```
