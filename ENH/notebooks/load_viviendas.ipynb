```python
file = '../2014/viviendas_enh_2014_csv.csv'
df_v14 = pd.read_csv(file, encoding='latin')
df_v14['year'] = 2014
df_v14.head()
```

```python
file = '../2015/vivienda.csv'
df_v15 = pd.read_csv(file, encoding='latin')
df_v15['year'] = 2015
df_v15.head()
```

```python
file = '../2016/vivienda.csv'
df_v16 = pd.read_csv(file, encoding='latin')
df_v16['year'] = 2016
df_v16.head()
```

```python
file = '../2017/vivienda.csv'
df_v17 = pd.read_csv(file, encoding='latin')
df_v17['year'] = 2017
df_v17.head()
```

```python
df = pd.concat([df_v14, df_v15, df_v16, df_v17])
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.columns[df_copy.isnull().any()].tolist()
```
