```python
import pandas as pd
```

```python
file = '../2014/hogares_enh_2014_csv.csv'
df_h14 = pd.read_csv(file, encoding='latin')
df_h14['year'] = 2014
df_h14.head()
```

```python
file = '../2015/hogar.csv'
df_h15 = pd.read_csv(file, encoding='latin')
df_h15['year'] = 2015
df_h15.head()
```

```python
file = '../2016/hogar.csv'
df_h16 = pd.read_csv(file, encoding='latin')
df_h16['year'] = 2016
df_h16.head()
```

```python
file = '../2017/hogar.csv'
df_h17 = pd.read_csv(file, encoding='latin')
df_h17['year'] = 2017
df_h17.head()
```

```python
df = pd.concat([df_h14, df_h15, df_h16, df_h17])
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.shape
```

```python
df_copy.columns[df.isnull().any()].tolist()
```

```python
df_copy.fillna(value=-1, axis=0, inplace=True)
```

```python
df_copy.replace({' ' : -1}, inplace=True)
```

```python
df_copy.replace({'1.79769313486232e+308' : -1}, inplace=True)
```

```python
df_copy.isnull().any().any()
```

```python
len(pd.isnull(df_copy).any(1).nonzero()[0].tolist())
```

```python
df_copy = df_copy.astype('int64')
```

```python
for line in df_copy.columns.tolist():
    #print(line)
    try:
        df_copy[line] = df_copy[line].astype('int64')
    except:
        print(line)
```

```python
for line in df_copy.columns:
    print(line, ' : ', df_copy[line].unique(), '\n')
```
