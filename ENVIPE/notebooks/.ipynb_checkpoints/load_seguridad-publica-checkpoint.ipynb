Características correspondientes a los delitos captados por el módulo sobre victimización.

```python
import pandas as pd
```

```python
file = '../2014/bd_envipe2014/bd_envipe2014/TMod_Vic.csv'
df_v14 = pd.read_csv(file, encoding='latin')
df_v14['year'] = 2014
df_v14.head()
```

```python
file = '../2015/TMod_Vic.csv'
df_v15 = pd.read_csv(file, encoding='latin')
df_v15['year'] = 2015
df_v15.head()
```

```python
file = '../2016/tmod_vic_envipe2016/conjunto_de_datos/tmod_vic_envipe2016.csv'
df_v16 = pd.read_csv(file, encoding='latin')
df_v16['year'] = 2016
df_v16.head()
```

```python
file = '../2017/tmod_vic_envipe2017/conjunto_de_datos/tmod_vic.csv'
df_v17 = pd.read_csv(file, encoding='latin')
df_v17['year'] = 2017
df_v17.head()
```

```python
df = pd.concat([df_v14, df_v15, df_v16, df_v17])
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.shape
```

```python
df_copy.columns[df.isnull().any()].tolist()
```

```python
df_copy.fillna(value=-1, axis=0, inplace=True)
```

```python
df_copy.replace({' ' : -1}, inplace=True)
```

```python
df_copy.replace({'\r' : -1}, inplace=True)
```

```python
df_copy.replace({99 : -1, '99' : -1}, inplace=True)
```

```python
df_copy.replace({'0\r' : 0,
                '1\r' : 1,
                '2\r' : 2,
                '3\r' : 3,
                '4\r' : 4,
                '5\r' : 5,
                '6\r' : 6,
                '7\r' : 7,
                '8\r' : 8,
                '9\r' : 9}, inplace=True)
```

```python
df_copy.isnull().any().any()
```

```python
len(pd.isnull(df_copy).any(1).nonzero()[0].tolist())
```

```python
df_copy = df_copy.astype('int64')
```

```python
for line in df_copy.columns.tolist():
    #print(line)
    try:
        df_copy[line] = df_copy[line].astype('int64')
    except:
        print(line)
```

```python
for line in df_copy.columns:
    print(line, ' : ', df_copy[line].unique(), '\n')
```

```python
df_copy.to_csv('../dataframes/victimizacion.csv')
```
